```python
##extract radial distribution function from Hd(k)
G_d_meth = []
for r in np.arange(0, 12.1, 0.01):
    temp = sum([i*j*np.sin(i*r)*0.1 for i, j in zip(k, meth_Hd)])/(2*np.pi**2*r*0.02976) + 1.
    G_d_meth.append(temp)

G_d_etha = []
for r in np.arange(0, 12.1, 0.01):
    temp = sum([i*j*np.sin(i*r)*0.1 for i, j in zip(k, etha_Hd)])/(2*np.pi**2*r*0.03096) + 1.
    G_d_etha.append(temp)
    
G_d_meth = G_d_meth[1:]
G_d_etha = G_d_etha[1:]
```

```python
##plot g(r)
r = np.arange(0, 12.1, 0.01)[1:] ##skip the first element, 0, it is not allowed to have 0 in denominator
plt.plot(r, np.array(G_d_etha) + 2.0 , "b-", lw=2, label = "ethanol")
plt.plot(r, G_d_meth, "r-", lw=2, label = "methanol")
plt.legend()
plt.xlim([0,12])
plt.xlabel("r/A")
plt.ylabel("gr")
```

```python
##estimate the peak positions of ethanol and methanol
r_selected = r[220:320]
G_d_meth_selected = G_d_meth[220:320]
G_d_etha_selected = G_d_etha[220:320]
est_meth_pos = r_selected[max(enumerate(G_d_meth_selected),key=lambda x: x[1])[0]]
est_etha_pos = r_selected[max(enumerate(G_d_etha_selected),key=lambda x: x[1])[0]]
print ("The rough peak position for methanol is at {} A".format(est_meth_pos))
print ("The rough peak position for ethanol  is at {} A".format(est_etha_pos))
```

```python
##define a Gaussian function
def gauss(r, r0, Gamma, scale):
    return scale*(np.sqrt(np.log(2.)/np.pi)*(2./Gamma)*np.exp((-1)*(r-r0)**2*4.*np.log(2.)/Gamma**2))
```

```python
plt.plot(r, gauss(r, 2, 0.5, 1), "r-", lw=2, label = "gauss_for_fun")
plt.xlim([0,4])
```

```python
##first methanol Gaussian fit
r_1 = r[220:300]
G_1 = G_d_meth[220:300]
from scipy.optimize import curve_fit
from scipy.integrate import simps
p0 = [2.79, 0.05, 0.3]
popt, pcov = curve_fit(gauss, r_1, G_1, p0)
print (popt)
plt.plot(r, G_d_meth, "b-", lw=2, label = "methanol_gr")
plt.plot(r_1, gauss(r_1, popt[0], popt[1], popt[2]), "r-", lw=2, label = "fit")
plt.legend()
print(simps(gauss(r_1, popt[0], popt[1], popt[2]), r_1))
```

```python
##second methanol Gaussian fit
r_2 = r[220:320]
G_2 = G_d_meth[220:320]
from scipy.optimize import curve_fit
p0 = [2.79, 0.05, 0.3]
popt, pcov = curve_fit(gauss, r_1, G_1, p0)
print (popt)
plt.plot(r, G_d_meth, "b-", lw=2, label = "ethanol_gr")
plt.plot(r_2, gauss(r_2, popt[0], popt[1], popt[2]), "r-", lw=2, label = "fit")
plt.legend()
print(simps(gauss(r_2, popt[0], popt[1], popt[2]), r_2))
```
